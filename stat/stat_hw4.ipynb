## Задача 1

Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800]. Найдите ее среднее значение и дисперсию.

### Решение

$f_{X}(x)=\left\{{\begin{matrix}{\dfrac {1}{b-a}},&x\in [a,b]\\0,&x\not \in [a,b]\end{matrix}}\right..$

$M(X)=\frac{a+b}{2}=\frac{1000}{2}=500$

$D(X)=\frac{(b-a)^2}{12}=\frac{600}{12}=50$

## Задача 2

О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
Если да, найдите ее.

### Решение

Решим уравнение

$D(X)=\frac{(b-0.5)^2}{12}=0.2$

$b^2-b+0.25-2.4=0$

$b_1 \approx -1.04919,\, b_2 \approx 2.04919$

Так как левая граница $a=0.5$, то правая граница $b=2.04919$

Найдем среднее значение:

$M(X)=\frac{a+b}{2}=\frac{2.54919}{2}=1.274595$

## Задача 3

Непрерывная случайная величина X распределена нормально и задана плотностью распределения
$$f(x) = \frac{1}{4\sqrt{2\pi}}e^{-\frac{(x+2)^2}{32}}.$$
Найдите:
а) M(X)
б) D(X)
в) std(X) (среднее квадратичное отклонение)

### Решение

$$f(x)={\frac {1}{\sigma {\sqrt {2\pi }}}}e^{-\frac {(x-a)^2 }{2\sigma^2}}$$

Следовательно,

$M(x)=-2, \, \sigma(x) = 4, \, D(x)=16$

## Задача 4

Рост взрослого населения города X имеет нормальное распределение.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:

а). больше 182 см

б). больше 190 см

в). от 166 см до 190 см

г). от 166 см до 182 см

д). от 158 см до 190 см

е). не выше 150 см или не ниже 190 см

ё). не выше 150 см или не ниже 198 см

ж). ниже 166 см.

### Решение

$Z(x)=\frac{x-\mu}{\sigma}$

**а). больше 182 см**

$Z(182)=\frac{182-174}{8}=1$

$N(Z)=0.8413$

Следовательно, > 182 см имеют рост $1 - 0.8413 = 0.1587$ жителей

**б). больше 190 см**

$Z(190)=\frac{190-174}{8}=2$

$N(Z)=0.9772$

Следовательно, > 190 см имеют рост $1 - 0.9772 = 0.0228$ жителей

**в). от 166 см до 190 см**

$Z(166)=\frac{166-174}{8}=-1$

$N(Z)=0.1587$

Следовательно, от 166 см до 190 см имеют рост $1 - 0.1587 - 0.0228 = 0.8185$ жителей

**г). от 166 см до 182 см**

От 166 см до 182 см имеют рост $1 - 0.1587 - 0.1587 = 0.6826$ жителей

**д). от 158 см до 190 см**

$Z(158)=\frac{158-174}{8}=-2$

$N(Z)=0.0228$

Следовательно, от 158 см до 190 см имеют рост $1 - 0.0228 - 0.0228 = 0.9544$ жителей

**е). не выше 150 см или не ниже 190 см**

$Z(150)=\frac{150-174}{8}=-3$

$N(Z)=0.0014$

Следовательно, рост не выше 150 см или не ниже 190 см имеют $0.0014 + 0.0228 = 0.0242$ жителей

**ё). не выше 150 см или не ниже 198 см**

$Z(198)=\frac{198-174}{8}=3$

$N(Z)=0.9986$

Следовательно, рост не выше 150 см или не ниже 198 см имеют $0.0014 + 1 - 0.9986 = 0.0028$ жителей

**ж). ниже 166 см**

Рост ниже 166 см имеют $0.1587$ жителей

## Задача 5

На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

### Решение

$\sigma(x) = \sqrt{25}=5$

Следовательно,

$Z(190)=\frac{190-178}{5}=\frac{12}{5}=2.4$
