## Задача 1

Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

### Решение

$n = 100,\, k = 85,\, p = 0.8,\, q=0.2$

Воспрозуемся формулой Бернулли:

$$P_{n}(X=k)=C_{n}^{k}p^{k}q^{n-k}$$

$C_{100}^{85}\cdot 0.8^{85}\cdot 0.2^{15}\approx 0.048 $


## Задача 2

Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?

### Решение

####1) Какова вероятность, что ни одна из них не перегорит в первый день?

$n = 5000,\, m = 0,\, p = 0.0004$

Воспользуемся формулой Пуассона:

$$P_m \approx =\frac{\lambda^m}{m!}e^{-\lambda},\,\, \lambda=n\cdot p$$

$\lambda=5000 \cdot 0.0004 = 2$

$P_m \approx =\frac{2^{0}}{0!}e^{-2}=1\cdot e^{-2}\approx 0.1353$

####2) Какова вероятность, что перегорят ровно две?

$n = 5000,\, m = 2,\, p = 0.0004$

Воспользуемся формулой Пуассона:

$\lambda=5000 \cdot 0.0004 = 2$

$P_m \approx =\frac{2^{2}}{2!}e^{-2}=2\cdot e^{-2}\approx 0.2707$


## Задача 3

Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?

### Решение

$n = 144,\, k = 70,\, p = 0.5,\, q=0.5$

Воспрозуемся формулой Бернулли:

$C_{144}^{70}\cdot 0.5^{70}\cdot 0.5^{74}=C_{144}^{70}\cdot 0.5^{144}\approx 0.06281 $


## Задача 4

В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?

### Решение

####1) Какова вероятность того, что все мячи белые?

Найдем вероятность того, что из первого ящика вытащили 2 белых мяча, и из второго ящика вытащили 2 белых мяча, и перемножим их.

а) $p_1=\frac{C_{7}^{2}}{C_{10}^{2}}=\frac{21}{45}$

б) $p_2=\frac{C_{9}^{2}}{C_{11}^{2}}=\frac{36}{55}$

$p=p_1\cdot p_2=\frac{21}{45}\cdot \frac{36}{55} \approx 0.305$

####2) Какова вероятность того, что ровно два мяча белые?

Найдем вероятность того, что из первого ящика вытащили 2 белых мяча, а из второго ни одного; из каждого ящика вытащили по одному белочу мячу; из первого ящика не вытащили ни одного белого мяча, а из второго 2. Найденные вероятности просуммируем.

а) $p_1=\frac{C_{7}^{2}}{C_{10}^{2}}=\frac{21}{45}$

$p_2=\frac{C_{2}^{2}}{C_{11}^{2}}=\frac{1}{55}$

$p'_1=p_1\cdot p_2=\frac{21}{45}\cdot \frac{1}{55} \approx 0.008485$

б) $p_1=\frac{C_{7}^{1}\cdot C_{3}^{1}}{C_{10}^{2}}=\frac{21}{45}$

$p_2=\frac{C_{9}^{1}\cdot C_{2}^{1}}{C_{11}^{2}}=\frac{18}{55}$

$p'_2=p_1\cdot p_2=\frac{21}{45}\cdot \frac{18}{55} \approx 0.152727$

в) $p_1=\frac{C_{3}^{2}}{C_{10}^{2}}=\frac{3}{45}$

$p_2=\frac{C_{9}^{2}}{C_{11}^{2}}=\frac{36}{55}$

$p'_3=p_1\cdot p_2=\frac{3}{45}\cdot \frac{36}{55} \approx 0.043636$

Ответ: $0.008485+0.152727+0.043636=0.204848\approx 0.2$

####3) Какова вероятность того, что хотя бы один мяч белый?

Вероятность равно 1 минус вероятность, что у нас нет ни одного белого мяча.

$p_1=\frac{C_{3}^{2}}{C_{10}^{2}}=\frac{3}{45}$

$p_2=\frac{C_{2}^{2}}{C_{11}^{2}}=\frac{1}{55}$

$p_3=p_1\cdot p_2=\frac{3}{45}\cdot \frac{1}{55} \approx 0.001212$

Ответ: $1-0.001212=0.998788$
